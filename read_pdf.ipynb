```python
import pdfplumber

# Path to your PDF file
pdf_path = 'pdfs/ip309.pdf'

with pdfplumber.open(pdf_path) as pdf:
    page = pdf.pages[42]
    if page:
    # for page in pdf.pages:
        # Define the dimensions of the page
        width = page.width
        height = page.height
        
        # Assume the column separator is approximately at the midpoint of the page width
        column_width = width / 2
        
        # Define bounding boxes for the left and right columns
        left_column = (0, 50, column_width, height - 50)
        right_column = (column_width, 50, width, height - 50)

        im = page.to_image(resolution=150)
        
        # Extract text from the left column
        left_text = page.within_bbox(left_column).extract_text()
        if left_text:
            print("Text from the left column:")
            print(left_text)
        
        im.draw_rects(page.within_bbox(left_column).extract_words())
        
        # Extract text from the right column
        right_text = page.within_bbox(right_column).extract_text()
        if right_text:
            print("Text from the right column:")
            print(right_text)
```
